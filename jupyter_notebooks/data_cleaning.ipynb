## Credit Card Churn Data Cleaning

```python
# import Pandas
import pandas as pd
import seaborn as sns
```

```python
# load the dataset and check the first few rows
df = pd.read_csv('../Data/Raw/BankChurners.csv')
df.head()
```

```python
#checking for missing values
df.isnull().sum()
```

```python
#checking data types
df.dtypes
```

There are two collumns that are unnecessary for analysis and it will be dropped.

```python
df = df.drop(labels=['Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2'], axis=1)
df = df.drop(labels=['Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1'], axis=1)
df.head(10)
```

```python
df.dtypes
```

```python
# checking for duplicates
dupes = df.duplicated()
dupes.sum()
```

```python
# Checking all the columns in the DataFrame
df.columns
```

```python
#renaming the CLIENTNUM column to Client_id to make it more readable and consistent
df.rename(columns={'CLIENTNUM': 'Client_id'}, inplace=True)
df.head()
```

```python
df['Education_Level'].unique()
```

```python
# Creating a new column 'Churn_Flag' based on the 'Attrition_Flag' column
# This will help in binary classification tasks where 0 indicates an existing customer and 1 indicates
df['Churn_Flag'] = df['Attrition_Flag'].map({
    'Existing Customer': 0,
    'Attrited Customer': 1
})

df.head()
```

```python
# saving the cleaned DataFrame to a new CSV file
df.to_csv('cleaned_data.csv', index=False)
```
